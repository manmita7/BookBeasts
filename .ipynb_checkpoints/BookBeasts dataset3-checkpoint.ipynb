```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#top100 is the top 100 books from 1998-2010
#Volume is how many books were sold which is a great thing to potentially track
#Nielsen records 1998-2010
top100_df = pd.read_csv('Resources/Top 100 books - TOP 100 - ALL TIME.csv')

#convert values from pounds to dollars
#remove pound sign
top100_df["Value"] = [x.strip('£') for x in top100_df["Value"]]
top100_df["RRP"] = [x.strip('£') for x in top100_df["RRP"]]
top100_df["ASP"] = [x.strip('£') for x in top100_df["ASP"]]
#convert str to float
top100_df["Value"] = [float(x.replace(',','')) for x in top100_df["Value"]]
top100_df["Value"] = pd.to_numeric(top100_df["Value"])
top100_df["RRP"] = pd.to_numeric(top100_df["RRP"])
top100_df["ASP"] = pd.to_numeric(top100_df["ASP"])
#convert to dollars
top100_df["Value"] = [x*1.22 for x in top100_df["Value"]]
top100_df["RRP"] = [x*1.22 for x in top100_df["RRP"]]
top100_df["ASP"] = [x*1.22 for x in top100_df["ASP"]]

#extract months and year
top100_df["Publ Date"] = pd.to_datetime(top100_df['Publ Date'])
top100_df['year'] = top100_df['Publ Date'].dt.year
top100_df['month'] = top100_df['Publ Date'].dt.month
top100_df.head()
```

```python
# Calculate book_category Percents
book_category = top100_df.groupby("Product Class")["Product Class"].count()
book_category
```

```python
book_category = top100_df.groupby("Product Class")["Product Class"].count().to_dict()
book_category 
```

```python
sizes = book_category.values()
labels = book_category.keys()
#explode = (0.1, 0, 0, 0,)
colors = None
labels
```

```python
# Build Pie Chart
plt.pie(sizes, labels=labels, colors=None, shadow=True, startangle=150, autopct="%1.1f%%")
plt.axis('auto')
plt.title('% of book category', fontsize=20)
# Save Figure
plt.savefig('book_category.png')
# Show Figure
plt.show()
```

```python
#Binding Format ( paperback vs hardcover ) pie chart
binding_format = top100_df.groupby("Binding")["Binding"].count().to_dict()
sizes = binding_format.values()
labels = binding_format.keys()
colors = None
```

```python
# Build Pie Chart
plt.pie(sizes, labels=labels, colors=None, shadow=True, startangle=150, autopct="%1.1f%%")
plt.axis('auto')
plt.title('Binding Format', fontsize=20)
# Save Figure
plt.savefig('binding_format.png')
# Show Figure
plt.show()
```

```python
# top100 analysis by quarter over a decade
# Create bins and bin labels
quarter_bins = [0, 3, 6, 9, 12]
quarter_labels = ["1st Quarter", "2nd Quarter", "3rd Quarter", "4th Quarter"]
# Create Data Series Into New Column Inside DataFrame
top100_df["Quarter"] = pd.cut(top100_df["month"], bins=quarter_bins, labels=quarter_labels)
top100_df
```

```python
# quarterly top100 pie chart
quarterly_top100 = top100_df.groupby("Quarter")["Quarter"].count().to_dict()
sizes = quarterly_top100.values()
labels = quarterly_top100.keys()
colors = None

# Build Pie Chart
plt.pie(sizes, labels=labels, colors=None, shadow=True, startangle=150, autopct="%1.1f%%")
plt.axis('auto')
plt.title('Quarterly Top100', fontsize=20)
# Save Figure
plt.savefig('quarterly_top100.png')
# Show Figure
plt.show()

```
